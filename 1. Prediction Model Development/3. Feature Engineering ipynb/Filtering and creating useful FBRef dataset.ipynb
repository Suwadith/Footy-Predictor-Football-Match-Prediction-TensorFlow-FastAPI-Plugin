### Combining all "FBRef Matchday" seasonal datasets 

```python
import pandas as pd
import glob
import os
import numpy as np

# setting the path for joining multiple files
files = os.path.join("C:\\Users\\suwad\\OneDrive - Middlesex University\\Desktop\\MDX\Modules\\8. CST4444 Individual Project\\06. Rsearch Datasets\\Source\\1. FBRef\\3. Match Stats 2011-2022\\", "*.csv")

# list of merged files returned
files = glob.glob(files)

# joining files with concat and read_csv
pd.set_option('display.max_columns', None)
df = pd.concat(map(pd.read_csv, files), ignore_index=True)
df
```
