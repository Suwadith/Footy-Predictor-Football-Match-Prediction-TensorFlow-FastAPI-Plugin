```python
def assign_win_loss_streak_points(season_df):
    
    for index, row in season_df.iterrows():
        
        home_team_form = row['HM1'] + row['HM2'] + row['HM3'] + row['HM4'] + row['HM5']
        away_team_form = row['AM1'] + row['AM2'] + row['AM3'] + row['AM4'] + row['AM5']
        
        home_team_points = 0
        away_team_points = 0
        
        for char in range(0, len(home_team_form)):
            if home_team_form[char] == 'W':
                home_team_points += 3
            elif home_team_form[char] == 'D':
                home_team_points += 1
            else:
                home_team_points += 0
                
        for char in range(0, len(away_team_form)):
            if away_team_form[char] == 'W':
                away_team_points += 3
            elif away_team_form[char] == 'D':
                away_team_points += 1
            else:
                away_team_points += 0
                
        
#         season_df.at[index,'HomeGames'] = home_team_form
#         season_df.at[index,'AwayGames'] = away_team_form
                
        
        season_df.at[index,'DiffFormPts'] = home_team_points - away_team_points
        season_df.at[index,'DiffPts'] = row['HTP'] - row['ATP']
        
        season_df.at[index,'HTGD'] = row['HTGS'] - row['HTGC']
        season_df.at[index,'ATGD'] = row['ATGS'] - row['ATGC']
        
        season_df.at[index,'DiffLP'] = row['HomeTeamLP'] - row['AwayTeamLP']
        
        if home_team_form[0:3] == 'WWW':
            season_df.at[index,'HTWinStreak3'] = 1
        else:
            season_df.at[index,'HTWinStreak3'] = 0
            
        if home_team_form == 'WWWWW':
            season_df.at[index,'HTWinStreak5'] = 1
        else:
            season_df.at[index,'HTWinStreak5'] = 0
            
        if home_team_form[0:3] == 'LLL':
            season_df.at[index,'HTLossStreak3'] = 1
        else:
            season_df.at[index,'HTLossStreak3'] = 0
            
        if home_team_form == 'LLLLL':
            season_df.at[index,'HTLossStreak5'] = 1
        else:
            season_df.at[index,'HTLossStreak5'] = 0    
            
            
        if away_team_form[0:3] == 'WWW':
            season_df.at[index,'ATWinStreak3'] = 1
        else:
            season_df.at[index,'ATWinStreak3'] = 0
            
        if away_team_form == 'WWWWW':
            season_df.at[index,'ATWinStreak5'] = 1
        else:
            season_df.at[index,'ATWinStreak5'] = 0
            
        if away_team_form[0:3] == 'LLL':
            season_df.at[index,'ATLossStreak3'] = 1
        else:
            season_df.at[index,'ATLossStreak3'] = 0
            
        if away_team_form == 'LLLLL':
            season_df.at[index,'ATLossStreak5'] = 1
        else:
            season_df.at[index,'ATLossStreak5'] = 0
        
        
assign_win_loss_streak_points(combined_season_df)

combined_season_df.head()
```
