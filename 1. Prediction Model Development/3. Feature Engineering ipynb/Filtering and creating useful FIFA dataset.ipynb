### Combining all "FIFA Player Ratings" seasonal datasets 

```python
import pandas as pd
import glob
import os

# setting the path for joining multiple files
files = os.path.join("C:\\Users\\suwad\\OneDrive - Middlesex University\\Desktop\\MDX\Modules\\8. CST4444 Individual Project\\06. Rsearch Datasets\\Source\\2. FIFA\\1. Player Stats 2015-2022\\", "*.csv")

# list of merged files returned
files = glob.glob(files)

# joining files with concat and read_csv
pd.set_option('display.max_columns', None)
df = pd.concat(map(pd.read_csv, files), ignore_index=True)
df
```

```python
new_df = df.loc[df['league_name'] == "English Premier League"]
new_df
```

```python
new_df = new_df.loc[:, new_df.columns.isin(['season', 'long_name', 'player_positions', 
                                    'overall', 'value_eur', 'wage_eur', 'age',
                                    'club_name', 'league_name', 'international_reputation', 
                                    'pace', 'shooting', 'passing', 'dribbling', 'physic'])]
new_df
```

```python
grouped_df = new_df.groupby(['club_name', 'season'])[['overall', 'age', 'value_eur', 'wage_eur', 
                                        'international_reputation', 'pace', 'shooting', 
                                        'passing', 'dribbling', 'physic']].sum()
grouped_df
```

```python
reset_df = grouped_df.reset_index()
reset_df
```

```python
reset_df.loc[reset_df['season'] == 15, 'season'] = 14
reset_df.loc[reset_df['season'] == 16, 'season'] = 15
reset_df.loc[reset_df['season'] == 17, 'season'] = 16
reset_df.loc[reset_df['season'] == 18, 'season'] = 17
reset_df.loc[reset_df['season'] == 19, 'season'] = 18
reset_df.loc[reset_df['season'] == 20, 'season'] = 19
reset_df.loc[reset_df['season'] == 21, 'season'] = 20
reset_df.loc[reset_df['season'] == 22, 'season'] = 21
```
